```python
import pandas as pd
import numpy as np
import requests
```

```python
park_data = pd.read_csv("Park_Database.csv")
```

```python
park_data
```

```python
park_df = park_data.drop(columns = ["streettype","streetno", "streetdir", "streetnm", "streetsuff", "prkstatus", "prktype", "citycounci", "bocc", "prkdate", "operdays", "leasedid", "resvation", "reservfac",  "numparking"])
```

```python
park_df
```

```python
property_data = pd.read_csv("Park_Property.csv")
```

```python
property_data
```

```python
property_df = property_data.drop(columns= ["OBJECTID","STATUS", "PROPERTY"])
```

```python
property_df
```

```python
merge_df = pd.merge(park_df, property_df, how="left", on=["PARK_ID", "PARK_ID"])
```

```python
merge_df
```

```python
merge_df = merge_df.rename(columns={'prkphone': 'Park Phone #'})
```

```python
merge_df = merge_df.rename(columns={'parkname': 'Park Name'})
```

```python
merge_df = merge_df.rename(columns={'address': 'Address'})
```

```python
merge_df = merge_df.rename(columns={'city': 'City'})
```

```python
merge_df = merge_df.rename(columns={'state': 'State'})
```

```python
merge_df = merge_df.rename(columns={'x': 'Longitude'})
```

```python
merge_df = merge_df.rename(columns={'y': 'Latitude'})
```

```python
merge_df = merge_df.rename(columns={'zip': 'Zip Code'})
```

```python
merge_df = merge_df.rename(columns={'park district': 'Park District'})
```

```python
merge_df.head()
```

```python
CleanPark_database = merge_df
```

```python
CleanPark_database
```
